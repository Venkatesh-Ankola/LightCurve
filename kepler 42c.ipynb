```python
from lightkurve import search_targetpixelfile
Kepler_42 = search_targetpixelfile("Kepler-42", quarter = 9).download()
```

```python
Kepler_42.plot();
```

```python
Kepler_42.plot_pixels()
```

```python
light_curve = Kepler_42.to_lightcurve(aperture_mask = Kepler_42.pipeline_mask)
light_curve
```

```python
light_curve.time , light_curve.flux
```

```python
light_curve.plot()
```

```python
flat_lc = light_curve.flatten(window_length=201)
flat_lc.plot();
```

```python
fold_lc = flat_lc.fold(period=301)
fold_lc.plot()
```
